```python
import numpy as np
from matplotlib import pyplot as plt

theta = np.linspace(-180.0, 180.0, num=(360 // 5) + 1)
theta = np.deg2rad(theta)
P_MU = np.load("P_MU_sample.npy")

plt.figure(figsize=(15, 10))
i_max = 5
j_max = 4
for i in range(i_max):
    for j in range(j_max):
        if i == 2 and j == 3:
            continue
        ax = plt.subplot(i_max, j_max, (j_max * i) + j + 1, projection="polar")
        ax.plot(theta, P_MU)
        ax.set_title("Spatial_Spectrum", fontsize=10)
        ax.set_theta_zero_location("N")
        ax.set_theta_direction(-1)
plt.show()
```

```python
theta = np.linspace(-180.0, 180.0, num=(360 // 5) + 1)
P_MU = np.load("P_MU_sample.npy")
plt.figure()
plt.plot(theta, P_MU, marker="o", linestyle="-", color="b")
plt.title("Spatial Spectrum")
plt.xlabel("Theta (degrees)")
plt.ylabel("P_MU Value")
plt.grid(True)
plt.xticks(np.arange(-180, 181, 20))
plt.yticks(np.linspace(np.min(P_MU), np.max(P_MU), 10))
plt.show()
```
