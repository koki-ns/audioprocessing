```python
import wave as wave
import numpy as np
import scipy.signal as sp

wav = wave

sample_wave_file = "./CMU_ARCTIC/cmu_us_aew_arctic/wav/arctic_a0001.wav"

wav = wave.open(sample_wave_file)

print(type(wav))

data = wav.readframes(wav.getnframes())
data = np.frombuffer(data, dtype=np.int16)
data = data/np.iinfo(np.int16).max

wav.close()

sample_rate = 16000

f, t, stft_data = sp.stft(data, fs=sample_rate, window="hann", nperseg=512, noverlap=256)

print(type(stft_data))
print("f:", f.shape) # 周波数インデックスの指す具体的な値が入っている
# print(f)
print("t:", t.shape) # 時間インデックスの指す具体的な値が入っている（mフレーム目の中心になる時刻らしい）
# print(t)
print("stft_data,shape: ", stft_data.shape)
print(stft_data[:, 0])
```
